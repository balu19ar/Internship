```python
#2) "Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
   # and make data frame"
    
page1 = requests.get('https://www.imdb.com/list/ls091520106/?sort=user_rating,desc&st_dt=&mode=detail&page=1')


soup = BeautifulSoup(page1.content)
soup

rating = []
for r in soup.find_all('div',class_='ipl-rating-star small'):
    rating.append(r.text.replace('\n',''))
rating

Imdb_titles = []
for im in soup.find_all('h3',class_='lister-item-header'):
    Imdb_titles.append(im.text.replace('\n',''))
Imdb_titles

Imdb_titles = [im.split('.')[1] for im in Imdb_titles]
Imdb_titles = [im.split('(')[0] for im in Imdb_titles]

Year_of_release = []
for y in soup.find_all('span',class_='lister-item-year text-muted unbold'):
    Year_of_release.append(y.text.replace('(I)',''))
Year_of_release


df = pd.DataFrame(
{
    'Imdb_titles' : Imdb_titles,
    'Year_of_release' : Year_of_release,
    'rating' : rating
})

df
```

# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame
                                                

```python
page2 = requests.get('https://www.imdb.com/list/ls056092300/')


soup = BeautifulSoup(page2.content)
soup

Imdb = []
for r in soup.find_all('h3',class_='lister-item-header'):
    Imdb.append(r.text.replace('\n',''))
Imdb

Imdb = [r.split('.')[1] for r in Imdb]
Imdb
Imdb = [r.split('(')[0] for r in Imdb]
Imdb

rate = soup.find_all(['div'], class_ = "ipl-rating-star small")
ratings = []
for i in rate:
    ratings.append(i.get_text().replace('\n',''))
ratings

year = []
for ye in soup.find_all('span',class_='lister-item-year text-muted unbold'):
    year.append(ye.text.replace('(I)',''))
year

df1 = pd.DataFrame({
    'Indian_Movies' : Imdb,
    'Rate' : ratings,
    'Year_of_release' : year
})

df1
```

# 4.Write s python program to display list of respected former presidents of India(i.e. Name , Term of office)


```python
page3 = requests.get('https://presidentofindia.nic.in/former-presidents.htm')


soup1 = BeautifulSoup(page3.content)


name = []
for n in soup1.find_all('h3'):
    name.append(n.text)

name = [n.split('(')[0] for n in name]


terms = []
for t in soup1.find_all('p'):
    terms.append(t.text)
terms = terms[0:16]


terms.remove('http://pranabmukherjee.nic.in',)

del terms[2]

del terms[3]
terms = [t.split(':')[1] for t in terms ]

df_president = pd.DataFrame(
    {
        'Name' : name,
        'Term Office' : terms
    })
df_president
```

# 9) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
dine = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dine

dine_soup = BeautifulSoup(dine.content)
dine_soup
```

```python
restaurant_name = []
for rn in dine_soup.find_all('a',class_='restnt-name ellipsis'):
    restaurant_name.append(rn.text)
restaurant_name
```

```python
cuisine = []
for c in dine_soup.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(c.text)
cuisine

cuisine = [c.split('|')[1] for c in cuisine]
cuisine
```

```python
locations = []
for loccc in dine_soup.find_all('div',class_='restnt-loc ellipsis'):
    locations.append(loccc.text)
locations
```

```python
ratings_D = []
for rd in dine_soup.find_all('div',class_='restnt-rating rating-4'):
    ratings_D.append(rd.text)
ratings_D
```

```python
images = []
for img in dine_soup.find_all('img',class_='no-img'):
    images.append(img['data-src'])
images
```

```python
df_dine = pd.DataFrame({
    'Restaurant_name' : restaurant_name,
    'Cuisine' : cuisine,
    'Location' : locations,
    'Ratings' : ratings_D,
    'Image_URL' : images
})

df_dine
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en i) Rank

ii) Publication

iii) h5-index

iv) h5-median

```python
google = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
google

google_soup = BeautifulSoup(google.content)
google_soup
```

```python
rank = []
for ra in google_soup.find_all('td',class_='gsc_mvt_p'):
    rank.append(ra.text)
rank

publication = []
for pub in google_soup.find_all('td',class_='gsc_mvt_t'):
    publication.append(pub.text)
publication

h5_index = []
for h5 in google_soup.find_all('a',class_='gs_ibl gsc_mp_anchor'):
    h5_index.append(h5.text)
h5_index

h5_median = []
for h5_ in google_soup.find_all('span',class_='gs_ibl gsc_mp_anchor'):
    h5_median.append(h5_.text)
h5_median

df_google = pd.DataFrame({
    'rank' : rank,
    'publication' : publication,
    'h5_index' : h5_index,
    'h5_median' : h5_median
})
df_google.head(6)
```
